### 4. Data Filtering and Indexing ( loc, iloc, where(),...)

```python
df = pd.read_excel("Excel_data.xlsx")
```

```python
# Query: List all records where salary > 10000

df[df['salary'] > 100000]
```

```python
# Query: List all records where salary > 100000 and slary < 150000

df[(df.salary > 100000) & (df.salary < 150000)]
```

```python
# Query: find the records where salary < 100000 or salary > 150000

df[(df.salary < 100000) | (df.salary > 150000)]
```

```python
# Query: find the record whose salary is maximum
df[df.salary == df.salary.max()]
```

```python
df[df.salary == df.salary.min()]
```

```python
# using where() to filter data
# where() keep the values where condition is satisfied but replace all other values with NaN
df1 = df.copy()
df1 = df1.where(df1.Age > 50)
df1
```

```python
df2 = df.copy()
df2.Age = df2.Age.where(df2.Age > 50,20)
df2
```

### Merging two dataframes

```python
df1 = pd.DataFrame({
    'ID': [1,2,3,4],
    'Hero': ['Ram Charan','Prabhas','Allu Arjun','Vijay']
})
df1
```

```python
df2 = pd.DataFrame({
    'ID': [2,3,4,5],
    'NetWorth': [100,200,150,80]
})
df2
```

```python
# merge() function merge the dataframes based upon common column
# merge() function behaves like the join funtion in SQL

merge_df = pd.merge(df1,df2, on  = 'ID')
merge_df
```

```python
merge_df = pd.merge(df1,df2, on  = 'ID', how = 'inner')
merge_df
```

```python
# left merge : Take all the data of left side table, but only common data from the right side table.

merge_df = pd.merge(df1,df2, on  = 'ID', how = 'left')
merge_df
```

```python
# right merge: This will taje complete data of right side table, but only common data from the left side table.

merge_df = pd.merge(df1,df2, on  = 'ID', how = 'right')
merge_df
```

```python
merge_df = pd.merge(df1,df2, on  = 'ID', how = 'outer')
merge_df
```

### concat() : used to combine dataframes ( Vertically(ROW-Wise) or Horizontally(Column-wise)

```python
df1 = pd.DataFrame({
    'ID' : [1,2],
    'Movie' : ['Son Of India', 'Kannappa']
})
df1
```

```python
df2 = pd.DataFrame({
    'ID' : [3,4],
    'Movie' : ['Mission Impossible', 'F1']
})
df2
```

```python
concat_df = pd.concat([df1,df2], axis = 0, ignore_index = True)  # axis = 0  menas Row wise
concat_df
```

```python
concat_df = pd.concat([df1,df2], axis = 1, ignore_index = True)  # axis = 1  menas Column wise
concat_df
```

```python
concat_df = concat_df.rename(columns = {0: 'ID', 1 : 'Tollywood', 2: 'ID2', 3: 'Hollywood'})
concat_df
                                
```

# Time series related Functions

```python
# Generating synthetic time series data    # Synthetic data = Artificial data

date_rng = pd.date_range(start = '2020-01-01', end = '2025-12-31',freq = 'h')
date_rng
```

```python
# Selecting 1000 records from above date time data

import numpy as np

np.random.seed(42)
random_dates = np.random.choice(date_rng, size = 1000, replace = False)
random_dates
```

```python
# Create a synthetic time series DataSet

df = pd.DataFrame({
    'timestamp' : pd.to_datetime(random_dates),
    'order_id' : np.arange(1,1001),
    'category' : np.random.choice(['Biryani', 'Paya', 'Irani Chai', 'Maska Bun'], size = 1000),
    'Quantity' : np.random.randint(1,5, size = 1000),
    'Price' : np.random.randint(25, 450, size = 1000)
})
df
```

```python
df['Sales'] = df['Price'] * df['Quantity']
```

```python
df
```

```python
df['Year'] = df.timestamp.dt.year
```

```python
df
```

```python
df['Month'] = df.timestamp.dt.month
```

```python
df
```

```python
df['Day'] = df.timestamp.dt.day
```

```python
df['Hour'] = df.timestamp.dt.hour
```

```python
df['Minute'] = df.timestamp.dt.minute
```

```python
df['Second'] = df.timestamp.dt.second
```

```python
df
```

```python
df.set_index('timestamp',inplace = True)
```
