## 3. Special type of numpy array creation

```python
# Array of zeros

# 1D Zero numpy array 
zero_arr = np.zeros(5)
zero_arr
```

```python
# 2D Zero numpy array 
zero_arr2 = np.zeros((3,3))
zero_arr2
```

```python
# 3D Zero numpy array 
zero_arr3 = np.zeros((2,2,3))
zero_arr3
```

```python
# Create and Initialize the numpy array with 1 (one)

# 1D Ones numpy array 
one_arr = np.ones(5)
one_arr 
```

```python
# 2D Ones numpy array 
one_arr2 = np.ones((2,3))
one_arr2
```

```python
# 3D Ones numpy array 
one_arr3 = np.ones((3,2,3))
one_arr3
```

```python
# filling/initialize the numpy array with specific value

# 1D array with specific value
spe_arr = np.full(5, 20000)
spe_arr
```

```python
# 2D array with specific value
spe_arr2 = np.full((3,2), 200)
spe_arr2
```

```python
# 3D array with specific value
spe_arr3 = np.full((3,3,2), 20)
spe_arr3
```

```python
# identity matrix

id = np.eye(4)
id
```

```python
# randint() : generating integer values in a fix range

rand_int_data = np.random.randint(10, 100, 10) # 10 is inclusive, 100 is exclusive
rand_int_data
```

```python
rand_int_data = np.random.randint(10, 100, (3,4)) # 10 is inclusive, 100 is exclusive
rand_int_data
```

```python
rand_int_data = np.random.randint(10, 100, (3,4,3)) # 10 is inclusive, 100 is exclusive
rand_int_data
```

```python
# choice() : This is used  for selecting categorical data

GF = np.array(['Turky', 'Murgi', 'Chuhiya', 'Bakri'])

np.random.choice(GF, size = 2)
```

```python
np.random.choice(GF, size = 2, replace = False)
```

```python
np.random.choice(GF, size = 10, p = [0.4, 0.3, 0.2, 0.1])
```

```python
np.random.seed(42)
np.random.choice(GF, size = 3)
```

```python
### Create sequences, shuffled, linearly, spaced array
arr1 = np.arange(1,10,1) # 1 is inclusive, 10 is exclusive
arr1
```

```python
np.random.shuffle(arr1)
arr1
```

```python
# permutation don't modify the original array
arr2 = np.random.permutation(arr1)
arr2
```

```python
# Generate linear spaced numbers between 1 to 10
arr3 = np.linspace(1,10,5) # linespace(start, end, size)
arr3
```
