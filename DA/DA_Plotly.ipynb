```python
import plotly.graph_objects as go
```

## 1. Surface Charts : Convex 3D Chart

```python
import numpy as np

# Data for 3D chart
x = np.linspace(-5,5,30)
y = np.linspace(-5,5,30)
x_new,y_new = np.meshgrid(x,y)    # Converting 1D data to 2D grid

z = x_new**2 + y_new**2

# Plot the 3D chart
fig = go.Figure(data = [go.Surface(x = x_new, y = y_new, z = z)]).update_layout(width = 800, height = 600)

fig.show()
```

## 2. Surface Chart (Non-convex 3D Chart)

```python
import numpy as np

# Data for 3D chart
x = np.linspace(-5,5,30)
y = np.linspace(-5,5,30)
x_new,y_new = np.meshgrid(x,y)    # Converting 1D data to 2D grid

z = np.sin(x_new) + np.cos(y_new)

# Plot the 3D chart
fig = go.Figure(data = [go.Surface(x = x_new, y = y_new, z = z)]).update_layout(width = 800, height = 600)

fig.show()
```

## 3. Scatter 3D Chart

```python
# Use : find the correleation 

np.random.seed(42)

x = np.random.randint(5, 20, 30)     # Number of Chocolates
y = np.random.randint(0, 25, 30)     # Number of replies
z = np.random.randint(5, 30, 30)     # Number of Gifts

fig = go.Figure(data = [go.Scatter3d(x = x, y = y, z = z, mode = 'markers')])

fig.update_layout(
    width = 800, 
    height = 600,
    scene = dict(
        xaxis_title = 'Number of Chocolates',
        yaxis_title = 'Number of replies',
        zaxis_title = 'Number of Gifts'
    )
)

fig.show()
```

## 4. Multi-Scatter 3D plot

```python
# Use : Shows Clustered Data


np.random.seed(42)

# Cluster-1 data 
x1 = np.random.randint(5, 20, 30)     # Number of Chocolates
y1 = np.random.randint(0, 25, 30)     # Number of replies
z1 = np.random.randint(5, 30, 30)     # Number of Gifts

# Cluster-2 data 
x2 = np.random.randint(10, 30, 30)     # Number of Chocolates
y2 = np.random.randint(0, 15, 30)     # Number of replies
z2 = np.random.randint(15, 40, 30)     # Number of Gifts

# Cluster-3 data
x3 = np.random.randint(15, 25, 30)     # Number of Chocolates
y3 = np.random.randint(0, 10, 30)     # Number of replies
z3 = np.random.randint(5, 40, 30)     # Number of Gifts


fig = go.Figure()

fig.add_trace(go.Scatter3d(x = x1, y = y1, z = z1, mode = 'markers'))
fig.add_trace(go.Scatter3d(x = x2, y = y2, z = z2, mode = 'markers'))
fig.add_trace(go.Scatter3d(x = x3, y = y3, z = z3, mode = 'markers'))


fig.update_layout(
    width = 800, 
    height = 600,
    scene = dict(
        xaxis_title = 'Number of Chocolates',
        yaxis_title = 'Number of replies',
        zaxis_title = 'Number of Gifts'
    )
)

fig.show()
```

## 3D Line Chart

```python
np.random.seed(42)

days = np.arange(1,31)   # Generates 1 to 30 values

x = days
y = np.sin(days / 3)
z = np.cos(days / 5)

fig = go.Figure(data = [go.Scatter3d(x = x, y = y, z = z, mode = 'lines')])

fig.update_layout(
    width = 800, 
    height = 600,
    scene = dict(
        xaxis_title = 'Number of Chocolates',
        yaxis_title = 'Number of replies',
        zaxis_title = 'Number of Gifts'
    )
)

fig.show()
```

## 3D Multi line plot

```python


np.random.seed(42)

# Cluster-1 data 
days = np.arange(1,31)   # Generates 1 to 30 values

x1 = days
y1 = np.sin(days / 3)
z1 = np.cos(days / 5)

# Cluster-2 data 
x2 = days
y2 = np.sin(days / 5)
z2 = np.cos(days / 3)

# Cluster-3 data
x3 = days
y3 = np.sin(days / 8)
z3 = np.cos(days / 5)


fig = go.Figure()

fig.add_trace(go.Scatter3d(x = x1, y = y1, z = z1, mode = 'lines',line = dict(width = 5)))
fig.add_trace(go.Scatter3d(x = x2, y = y2, z = z2, mode = 'lines'))
fig.add_trace(go.Scatter3d(x = x3, y = y3, z = z3, mode = 'lines'))


fig.update_layout(
    width = 800, 
    height = 600,
    scene = dict(
        xaxis_title = 'Number of Chocolates',
        yaxis_title = 'Number of replies',
        zaxis_title = 'Number of Gifts'
    )
)

fig.show()
```
