```python
# Mean

lie_test_data = np.array([3, 7, 0, 20, 2, 5, 30, 10])

mean_value = sum(lie_test_data) / len(lie_test_data)

print(mean_value)
```

## Measures of Dispersion (Spreadness of Data)

## MAD, Variance, SD

```python
# 1. Mean Absolute deviation (MAD):
#                MAD = (1/n) Σ |xi − x̄|         ( Σ i = 1 to n)    x̄ = mean



Reels_Watch_Count = np.array([50, 10, 20, 30, 40, 0, 0, 50, 0, 20, 40])

mean = np.mean(Reels_Watch_Count)

mean_abs = np.abs(Reels_Watch_Count - mean)

MAD = sum(mean_abs) / len(mean_abs)
print(MAD)
```

```python
# 2. Variance:
#                σ² = (1/n) Σ (xi − x̄)²         ( Σ i = 1 to n)     


print(np.var(Reels_Watch_Count))
```

```python
# 3. Standard Deviation (SD):
#                σ = √[(1/n) Σ (xi − x̄)²]       ( Σ i = 1 to n)     

print(np.std(Reels_Watch_Count))
```

```python
# Calculation LB, UB, Q1, Q2, Q3, Q4, IQR

Reels_Watch_Count = np.array([50, 10, 20, 30, 40, 0, 0, 50, 0, 20, 40])

Q1 = np.percentile(Reels_Watch_Count, 25)
print(Q1)

Q2 = np.percentile(Reels_Watch_Count, 50)
print(Q2)

Q3 = np.percentile(Reels_Watch_Count, 75)
print(Q3)

Q4 = np.percentile(Reels_Watch_Count, 100)
print(Q4)

IQR = Q3 - Q1
print(IQR)

LB = Q1 - 1.5 * IQR
print(LB)

UB = Q3 + 1.5 * IQR
print(UB)
```

# Covariance

```python
height = [150, 160, 170]
weight = [50, 55, 65]

mean_height = np.mean(height)
mean_weight = np.mean(weight)

print(mean_height)
print(mean_weight)
```

```python
height_update = height - mean_height
print(height_update)

weight_update = weight - mean_weight
print(weight_update)
```

```python
mul_height_weight = height_update * weight_update
print(mul_height_weight)
```

```python
pop_Cov = sum(mul_height_weight) / len(mul_height_weight)
print(pop_Cov)
```

```python
samp_Cov = sum(mul_height_weight) / (len(mul_height_weight)-1)
print(samp_Cov)
```

```python
# Covariance by built-in function

height = [150, 160, 170]
weight = [50, 55, 65]

cov_matrix = np.cov(height, weight, bias = True)        # Bias value decide to calculate population covariance or sample covareance
print(cov_matrix)
```

```python
cov_matrix = np.cov(height, weight, bias = False)
print(cov_matrix)
```

# Cosine Similarity.

# Cosine Distance

```python
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances
```

```python
# Boys vs Girls
A = [[1, 2, 3, 0, 1, 1, 10]]     # Boys data
B = [[30, 5, 3, 4, 6, 10, 20]]   # Girls data

similarity = cosine_similarity(A, B)
print(similarity)

distance = cosine_distances(A, B)
print(distance)
```

```python
# Application : Comparing text

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
text1 = 'I will start study form tomorrow, Promise'
text2 = 'I will not watch reels from tomorrow, Promise'
text3 = 'Immediatly he received his crush message, resoultions END'

vect_obj = TfidfVectorizer()
vectors = vect_obj.fit_transform([text1, text2, text3])
# print(vectors)

similarity = cosine_similarity(vectors)
print(similarity)

distance = cosine_distances(vectors)
print('cosine distance\n',distance)
```

# Probability Distribution

### DISCRETE DISTRINUTION



#### → Bernoulli Trail:

```python
prob = .30    # Probability of success

reply = np.random.binomial(n = 1, p = prob)

if reply == 1:
    print('Hurray, I am very happy')
else:
    print('She ignored the message')
```

####  → Binomial Distribution (Multiple Trails):

```python
total_msg_sent = 500
prob = .10


num_of_replies = np.random.binomial(n = total_msg_sent, p = prob)


print(num_of_replies)
```

#### → Poisson Distribution

```python
avg_reply = 10

reply_today = np.random.poisson(lam = avg_reply)

print(reply_today)
```

# Skewness

```python
# Generate sythetic data
np.random.seed(42)

data = {
    'Normal' : np.random.normal(loc = 0, scale = 1, size = 1000),         # Loc : mean, scale : SD
    'Positively_skewed' : np.random.exponential(scale = 2, size = 1000),  # scale : mean
    'Negatively_skewed' : -np.random.exponential(scale = 2, size = 1000),  # scale : mean
    'Uniform' : np.random.uniform(-3,3,1000)
}

df = pd.DataFrame(data)

print('Skewness')
print(df.skew())

print("Kurtosis")
print(df.kurtosis())
```

```python
plt.figure(figsize = (20, 15))
for i, col in enumerate(df.columns):
    plt.subplot(2,2,i+1)
    sns.histplot(df[col], kde = True)
    plt.title(f'{col}\n skewn: {df[col].skew():.2f} | Kurt : {df[col].kurtosis():.2f}')
```

## Log Normal Transformation

```python
np.random.seed(42)

right_skewed_data = np.random.exponential(scale = 2, size = 1000)

transformed_data = np.log(right_skewed_data + 0.2)

plt.subplot(1,2,2)
sns.histplot(transformed_data, kde =)

plt.subplot(1,2,1)
sns.histplot(right_skewed_data)
```

### Left skewed to normal distributed

```python
from sklearn.preprocessing import PowerTransformer
```

```python
np.random.seed(42)
left_skewed_data = -np.random.exponential(scale = 2, size = 1000)

pt = PowerTransformer(method = 'yeo-johnson')
transformed_Data = pt.fit_transform(left_skewed_data.reshape(-1,1))

plt.subplot(1,2,1)
sns.histplot(left_skewed_data)

plt.subplot(1,2,2)
sns.histplot(transformed_Data)
```
