### 'flights' Dataset

```python
df1 = sns.load_dataset('flights')
df1
```

## Line chart : Trends over the time

```python
# Find the ternds of number of passengers travels per year

timeline = df1.groupby('year')['passengers'].sum().reset_index()
timeline
```

```python
sns.lineplot(x = 'year', y = 'passengers', data = timeline, marker = 'o')
plt.title('Year-wise passengers detail')
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = df)
```

### Pair plot

```python
# Use : A Pair plot is a grid of scatter plot between all numerical variables

sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

### Heatmap

```python
# Heatmap helps to visualize correleation between many numerical varialbles/features.

# Why we are finding the correleation between variables?
# so that we can find the highlly correleating features.
# In ML, we must remove one of the highly correleated feature to make ML works better.


# Correleation Matrix : correleation betweeen many numerical variables/features.

corr_mat = df.corr(numeric_only = True)
corr_mat
```

```python
# Heatmap is nothing but just the better representaion of correlation mattrix

sns.heatmap(corr_mat, annot = True, cmap = 'coolwarm')
```

# Categorical feature releation : pivot table + heatmap

```python
df
```

```python
# Pivot table: Summary table of DF(Data Frame)
# aggfun = size : count of male & female on particular day

pivot = df.pivot_table(index = 'day', columns = 'sex', aggfunc = 'size', observed=False)  # observed=False to remove warning
print(pivot)
```

```python
sns.heatmap(pivot, annot = True, cmap = 'coolwarm')
```

```python
# Example-2

pivot = df.pivot_table(index = 'day', columns = 'sex', values = 'total_bill', aggfunc = 'mean', observed=False)  # observed=False to remove warning
sns.heatmap(pivot, annot = True, cmap = 'coolwarm')
```

```python
# Example-3

pivot = df.pivot_table(index = ['day', 'time'], columns = ['sex','smoker'], aggfunc = 'size', observed=False)  # observed=False to remove warning
sns.heatmap(pivot, annot = True, cmap = 'coolwarm')
```

# countplot() : Shows Frequency of Categorical data

```python
sns.countplot(data = df, x = 'smoker', hue = 'sex')
```

### Box plot

```python
# Use: Find Outliers & Skewness

# Example-1

sns.boxplot(y = 'tip', data = df)
```

```python
# Example-2

sns.boxplot(x = 'day', y = 'tip', data = df)
```

```python
# Example-3

sns.boxplot(y = 'day',x = 'tip', data = df, hue = 'smoker')
```

```python
# Example - 4 : plot multiple numerical columns

sns.boxplot(data = df[['total_bill','tip']])
```

## Density plot (KDE plot) - Kernal Density Estimate

```python
# Use : KDE plot is used to visualize probablility of density function (PDF)
# PDF : This represents probability over a inrerval(range)

# Example : find the range of total_bill which max people is paying

sns.kdeplot(data = df, x = 'total_bill', fill = True)
```

## Violin chart

```python
# Violin Chart = combination of Box plot + KDE Plot

sns.violinplot(x = 'day', y = 'total_bill', data = df)
```

## Regression Plot

```python
# Use: This will plot the "Simple Linear Regression MLL' modal

sns.regplot(x = 'total_bill', y = 'tip', data = df)
```

## facetfrid plot

```python
# Use : it is used for creating multiple subplots based upon sub-classes in categorical variable

g = sns.FacetGrid(df, col = 'sex')
g.map_dataframe(sns.scatterplot, x = 'total_bill', y = 'tip')
```

## joinplot

```python
# Use: it is used for finding releationship between two numerical variables
# It also shows the individual distribution of data

sns.jointplot(data = df, x = 'total_bill', y = 'tip', kind = 'scatter')
```
