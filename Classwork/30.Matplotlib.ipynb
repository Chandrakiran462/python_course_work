```python
import matplotlib.pyplot as plt
hours = [1,2,3,4,5,6,7,8]
visitors = [50,70,60,80,40,100,120,90]

plt.plot(hours,visitors,
         color = 'red',
         marker = 'o',
         linestyle = '-.',
         linewidth = 1,
         label = 'visitors')
plt.title("Web site visitors")
plt.xlabel('Hours')
plt.ylabel('visitors per hour')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

ad_spent = [5, 25, 28, 22, 27, 8, 23, 6, 24, 21, 17, 26, 9, 19, 16, 18, 7, 10, 20, 13, 14, 29, 11, 15, 12]
sales = [53, 47, 88, 92, 78, 55, 54, 91, 82, 48, 87, 44, 90, 74, 83, 75, 63, 59, 71, 49, 89, 86, 70, 50, 66]

sizes = list(map(lambda i : i*4,sales))

colors = sales

plt.scatter(ad_spent,sales,
            s = sizes,
            c = colors,
            cmap = "viridis",
            alpha = 0.8,
            edgecolors="black",
            linewidths=1
            )

plt.title("Ads Spent vs sales")
plt.xlabel("Ads_spent")
plt.ylabel("Sales")
plt.colorbar(label = "Sales Value")
plt.grid(True,linestyle = "--",alpha = 0.3)
plt.show()
```

```python
brands = ["Apple","Samsung","Xiaomi","Oppo","Vivo","Others"]
market_share = [15,18,22,17,18,10]

plt.pie(market_share,
        labels = brands,
        autopct = "%1.1f%%",
        startangle = 270,
        colors = ["red","green","blue","yellow","orange","brown"],
        explode = [0,0,0.1,0,0,0],
        shadow = True
)

plt.title("Mobile market share")
plt.show()
```
