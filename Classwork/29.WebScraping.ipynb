```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://timely-sunshine-e821b3.netlify.app'
```

```python
page = requests.get(URL)
```

```python
page.status_code
```

```python
htmlCode = page.text
print(htmlCode)
```

```python
soup = BeautifulSoup(htmlCode,"html.parser")
print(soup)
```

```python
content = soup.find('div',attrs = {'class':'a'})
print(content)
```

```python
content = soup.find_all('div',attrs = {'class':'a'})
for i in content:
  print(i,end = "\n\n\n\n\n\n\n\n")
```

```python
content = soup.find('div',attrs = {'class':'name'})
text = content.strip()
print(text)
```

```python
content = soup.find('div',attrs = {'class':'image'})
print(content)
```

```python
img_tag = content.find('img')
print(img_tag)
```

```python
img_src = img_tag['src']
print(img_src)
```

```python
items = soup.find_all('div',attrs = {'class':'a'})
names = []
prices = []
images = []

for item in items:
  name = item.find('div',class_ = 'name').text.strip()
  price = item.find('div',class_ = 'price').text.strip()
  img_tag = item.find('div',class_ = 'image').find('img')
  img_src = img_tag['src'] if img_tag else None

  names.append(name)
  prices.append(price)
  images.append(img_src)

print("Names: ",names)
print("Prices: ",prices)
print("Images: ",images)
```
