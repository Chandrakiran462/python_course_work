```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("sales_data.csv")

# Assuming revenue is just the sum of 'Price' column
brand_revenue = df.groupby("brand_name")["price"].sum().reset_index()

# Plot line graph
plt.figure(figsize=(10,6))
plt.plot(brand_revenue["brand_name"], brand_revenue["price"], marker='o', linestyle='-')

# Add labels and title
plt.xlabel("Brand")
plt.ylabel("Revenue")
plt.title("Brand vs Revenue")
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```
